```python
import h5py
import matplotlib.pyplot as plt
import numpy as npm
```

```python
with h5py.File('known_B_E.h5', 'r') as f:
    #print('Keys:', list(f.keys()))
    Bx, By, Bz = f['Bx'][:], f['By'][:], f['Bz'][:]
    Ex, Ey, Ez = f['Ex'][:], f['Ey'][:], f['Ez'][:]
    Bnewx, Bnewy, Bnewz = f['Bnewx'][:], f['Bnewy'][:], f['Bnewz'][:]

    plt.plot(Ey, label=r'known $E_y=\cos(x)$')
    plt.plot(Ez, label=r'known $E_z=\sin(x)$')
    plt.plot(By, label=r'known $B_y=2$')
    plt.plot(Bz, label=r'known $B_z=-1$')
    plt.plot(Bnewy, label=r'$B^{new}_y=B_y-\partial E_z/\partial x$')
    plt.plot(Bnewz, label=r'$B^{new}_z=B_z+\partial E_y/\partial x$')
    plt.xlabel('x')
    plt.ylabel('field values')
    plt.legend()
    plt.show()
```
