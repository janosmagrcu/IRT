```python
import h5py
import matplotlib.pyplot as plt
import numpy as npm
```

```python
with h5py.File('sine_B.h5', 'r') as f:
    #print('Keys:', list(f.keys()))
    Bx, By, Bz = f['Bx'][:], f['By'][:], f['Bz'][:]
    Jx, Jy, Jz = f['Jx'][:], f['Jy'][:], f['Jz'][:]

    plt.plot(By, label=r'known $B_y=\cos(x)$')
    plt.plot(Bz, label=r'known $B_z=\sin(x)$')
    plt.plot(Jy, label=r'$J_y=-\partial B_z/\partial x$')
    plt.plot(Jz, label=r'$J_z=+\partial B_y/\partial x$')
    plt.xlabel('x')
    plt.ylabel('field values')
    plt.legend()
    plt.show()
```
